```python
import matplotlib.pyplot as plt
```

```python
prestroke_mrs = 1
untreated_mrs = 3.46
treated_mrs = 2.45

untreated_risk_mrs_5_6 = 0.27
treated_risk_mrs_5_6 = 0.1
```

```python
mrs_text = [
    'No disability',
    'No significant\ndisability',
    'Slight disability:\nNo assistance required',
    'Mild-Moderate disability:\nHelp required but\nable to walk unassisted',
    'Moderate disability:\nHelp required and\nunable to walk unassisted',
    'Severe disability:\nRequires much help\nand frequent assistance',
    'Dead'
]
```

```python
with plt.xkcd():

    fig = plt.figure(figsize=(9, 6))
    ax = fig.add_subplot(111)

    ax.set_xlim(-0.9, 6.6)
    ax.set_ylim(0, 0.77)

  

    ax.plot([0, 6], [0.6, 0.6], color='black', linewidth=0.5)

    # Remove axes and borders
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

    # Remove y axis ticks and labels
    ax.yaxis.set_ticks_position('none')
    ax.set_yticklabels([])
    # Remove x axis ticks and labels
    ax.xaxis.set_ticks_position('none')
    ax.set_xticklabels([])


    # Add mrs_text to the plot as text at X axis values 0 to 6
    for i, txt in enumerate(mrs_text):
        odd_i = i % 2 == 1
        y_offset = 0.1 + (0.12 * odd_i)

        # Add line to y = 0.6
        ax.plot([i, i], [0.6 - y_offset, 0.6], color='black', linewidth=0.5) 
        # Add text
        ax.text(i, 0.6 - y_offset, txt, ha='center', va='top', fontsize=9, color='black', bbox=dict(facecolor='white', edgecolor='white'))
        # Add value of i at y = 0.6
        ax.text(i, 0.59, i, ha='center', va='bottom', fontsize=9, color='black', bbox=dict(facecolor='white', edgecolor='white'))


    # Add labels to arrows
    ax.text(prestroke_mrs - 0.07, 0.65, 'Pre-stroke', ha='right', va='center', fontsize=9, color='blue', bbox=dict(facecolor='white', edgecolor='white'))
    ax.text(untreated_mrs, 0.65, 'Untreated', ha='left', va='center', fontsize=9, color='blue', bbox=dict(facecolor='white', edgecolor='white'))
    ax.text(prestroke_mrs - 0.07, 0.70, 'Pre-stroke', ha='right', va='center', fontsize=9, color='red', bbox=dict(facecolor='white', edgecolor='white'))
    ax.text(treated_mrs, 0.7, 'Treated', ha='left', va='center', fontsize=9, color='red', bbox=dict(facecolor='white', edgecolor='white'))

    # Draw arrows
    ax.annotate('', xy=(untreated_mrs, 0.65), xytext=(prestroke_mrs, 0.65), arrowprops=dict(edgecolor='blue', linewidth=3, arrowstyle='->'))
    ax.annotate('', xy=(treated_mrs, 0.70), xytext=(prestroke_mrs, 0.70), arrowprops=dict(edgecolor='red', linewidth=3, arrowstyle='->'))

    

    # Add text for treated_risk_mrs_5_6 and untreated_risk_mrs_5_6
    txt = f'Risk of death or severe disability:'
    ax.text(3.8, 0.2, txt, ha='left', va='bottom', fontsize=9, color='black', bbox=dict(facecolor='white', edgecolor='white'), fontweight='bold')
    risk = untreated_risk_mrs_5_6 * 100
    ax.text(3.8, 0.15, f'Untreated: {risk:0.0f} in 100 similar patients', ha='left', va='bottom', fontsize=9, color='black', bbox=dict(facecolor='white', edgecolor='white'))
    risk = treated_risk_mrs_5_6 *100
    ax.text(3.8, 0.1, f'Treated : {risk:0.0f} in 100 similar patients', ha='left', va='bottom', fontsize=9, color='red', bbox=dict(facecolor='white', edgecolor='white'))
    
    # Draw a box around the text
    ax.plot([3.7, 3.7], [0.08, 0.25], color='black', linewidth=0.5)
    ax.plot([6.3, 6.3], [0.08, 0.25], color='black', linewidth=0.5)
    ax.plot([3.7, 6.3], [0.08, 0.08], color='black', linewidth=0.5)
    ax.plot([3.7, 6.3], [0.25, 0.25], color='black', linewidth=0.5)

    # Add title
    ax.set_title('Typical outcome (using the modified Rankin Scale of disability)', fontsize=14, color='black')

    fig.show();
```
