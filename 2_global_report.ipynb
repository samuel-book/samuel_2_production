```python
from IPython.display import Image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
patient_predictions = pd.read_csv('./output/thrombolysis_choice_test_predictions.csv')
# Plot a histogram of the predicted probabilities
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
ax.set_title('Histogram of predicted probabilities')
ax.set_xlabel('Predicted probability of thrombolysis')
ax.set_ylabel('Frequency')
ax.hist(patient_predictions['probability'], bins=20)
# Add cumulative histogram counts
ax.hist(patient_predictions['probability'], bins=20, cumulative=True, 
        histtype='step', color='r', linewidth=1.5)
ax.grid()
plt.show()
```

```python
for prob in [0.25, 0.5, 0.75]:
    proportion = np.mean(patient_predictions['probability'] >= prob)
    print(f'{prob} : {proportion:0.3f}')


proportion = np.mean((
    patient_predictions['probability'] >= 0.25) &
    (patient_predictions['probability'] <= 0.75))
print(f'0.25-0.75 : {proportion:0.3f}')
```
